```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Normalfordeling

```python
my = 0 # gjennomsnitt
sigma = 0.5 # standardavvik
norm_dist = np.random.normal(my, sigma, 750)
sns.swarmplot(data= norm_dist)
plt.xticks([])
plt.ylabel("Verdier")
plt.show()
```

```python
sns.histplot(data=norm_dist, kde=True, bins=20)
plt.xlabel("Verdi")
plt.ylabel("Antall")
plt.show()
```

```python
my = 0 # gjennomsnitt
sigma = 0.5 # standardavvik
for n in np.logspace(2,7,6):
    norm_dist = np.random.normal(my, sigma, int(n))
    sns.histplot(data=norm_dist, kde=True, bins=20)
    plt.xlabel("Verdi")
    plt.ylabel("Antall")
    plt.show()
```
